```python
def identity(x):
    return x

class TextSelector(BaseEstimator, TransformerMixin):
    """Use on text columns in the data"""
    def __init__(self, key):
        self.key = key

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X[self.key]


class NumberSelector(BaseEstimator, TransformerMixin):
    """Use on numeric columns in the data"""

    def __init__(self, key):
        self.key = key

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X[[self.key]]
```

```python
def generate_feats_gender():
    char_ngrams = Pipeline([
        ('selector', TextSelector(key='lyrics')),
        ('tfidf', TfidfVectorizer(ngram_range=(5, 5), analyzer='char_wb', min_df=10,lowercase=False))
    ])
    word_ngrams = Pipeline([
        ('selector', TextSelector(key='lyrics')),
        ('tfidf', TfidfVectorizer(ngram_range=(1, 1), analyzer='word', min_df=5,lowercase=False))
    ])
    
    phonetic_char_ngrams = Pipeline([
        ('selector', TextSelector(key='phonetic_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(6, 7), analyzer='char_wb', min_df=10,lowercase=False))
    ])
    
    phonetic_word_ngrams = Pipeline([
        ('selector', TextSelector(key='phonetic_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(3, 6), analyzer='char_wb', min_df=10,lowercase=False))
    ])
    
    metaphone_char_ngrams = Pipeline([
        ('selector', TextSelector(key='metaphone_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(1, 4), analyzer='char', min_df=10,lowercase=False))
    ])
    
    soundex_char_ngrams = Pipeline([
        ('selector', TextSelector(key='soundex_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(2, 5), analyzer='char', min_df=10,lowercase=False))
    ])
    
    freqeuency_repr = Pipeline([
        ('selector', TextSelector(key='soundex_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(4, 8), analyzer='char', min_df=10,lowercase=False))
    ])
    
    length_repr_ngrams = Pipeline([
        ('selector', TextSelector(key='length_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(2, 3), analyzer='char', min_df=10,lowercase=False))
    ])
    
    punctC_repr_ngrams = Pipeline([
        ('selector', TextSelector(key='punctC_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(4, 8), analyzer='char', min_df=10,lowercase=False))
    ])
    
    vowel_repr_char_ngrams = Pipeline([
        ('selector', TextSelector(key='vowel_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(4, 9), analyzer='char_wb', min_df=10,lowercase=False))
    ])
    
    shape_repr_word_ngrams = Pipeline([
        ('selector', TextSelector(key='shape_repr')),
        ('tfidf', CountVectorizer(ngram_range=(4, 9), analyzer='char', min_df=10,lowercase=False))
    ])
    
    vowel_repr_word_ngrams = Pipeline([
        ('selector', TextSelector(key='vowel_repr')),
        ('tfidf', CountVectorizer(ngram_range=(1, 1), analyzer='word', min_df=10,lowercase=False))
    ])
    
    alliteration_repr = Pipeline([
        ('selector', TextSelector(key='alliteration_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(3, 10), analyzer='char'))
    ])
    
    ner_repr = Pipeline([
        ('selector', TextSelector(key='ner_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(1,3), analyzer='word'))
    ])
    
    pos_repr = Pipeline([
        ('selector', TextSelector(key='pos_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(3,6), analyzer='char', min_df=10,lowercase=False))
    ])
    
    sylla_repr = Pipeline([
        ('selector',TextSelector(key='syllab_repr')),
        ('tfidf', TfidfVectorizer(ngram_range=(2,3), analyzer='char', min_df=10,lowercase=False))
    ])
    
    word_count = Pipeline([
        ('selector', NumberSelector(key='word_count')),
        ('standard', StandardScaler())
    ])
    
    sentence_count = Pipeline([
        ('selector', NumberSelector(key='sentence_count')),
        ('standard', StandardScaler())
    ])
    
    exclam_count = Pipeline([
        ('selector', NumberSelector(key='!_count')),
        ('standard', StandardScaler())
    ])
    
    quest_count = Pipeline([
        ('selector', NumberSelector(key='?_count')),
        ('standard', StandardScaler())
    ])
    
    comma_count = Pipeline([
        ('selector', NumberSelector(key=',_count')),
        ('standard', StandardScaler())
    ])
    
    uniq_word_ratio = Pipeline([
        ('selector', NumberSelector(key='unique_word_ratio')),
        ('standard', StandardScaler())
    ])
    
    comma_ratio = Pipeline([
        ('selector', NumberSelector(key=',_ratio')),
        ('standard', StandardScaler())
    ])
    
    rep_sent_count_ratio = Pipeline([
        ('selector', NumberSelector(key='repeated_sentence_count_ratio')),
        ('standard', StandardScaler())
    ])
    
    rep_sent_ratio =Pipeline([
        ('selector', NumberSelector(key='repeated_sentence_count_ratio')),
        ('standard', StandardScaler())
    ])
    
    feats = [
        ('char_wb_ngrams', char_ngrams),
        #('word_ngrams',word_ngrams),
        ('phonetic_char_ngrams',phonetic_char_ngrams),
        ('metaphone_char_ngrams',metaphone_char_ngrams),
        ('soundex_char_ngrams',soundex_char_ngrams),
        ('length_repr_ngrams',length_repr_ngrams),
        ('punctC_repr_ngrams',punctC_repr_ngrams),
        ('shape_repr_word_ngrams',shape_repr_word_ngrams),
        ('vowel_repr_char_ngrams',vowel_repr_char_ngrams),
        ('vowel_repr_word_ngrams',vowel_repr_word_ngrams),
        ('alliteration_repr',alliteration_repr),
        #('ner_repr',ner_repr),
        ('pos_repr',pos_repr),
        #('syllable_repr',sylla_repr),
        ('word_count', word_count),
        ('sentence_count',sentence_count),
        #(',_count', comma_count),
        #('!_count', exclam_count),
        #('?_count',quest_count),
        ('unique_word_ratio',uniq_word_ratio),
        #('comma_ratio',comma_ratio),
        ('repeated_sentence_count_ratio',rep_sent_count_ratio),
        ('repeated_sentence_ratio',rep_sent_ratio)
    ]

    return feats

def return_feat_names():
    feat_names = [
        ('lexical_ngrams',TextSelector(key='lyrics')),
        ('phonetic_ngrams',TextSelector(key='phonetic_repr')),
        ('metaphone_ngrams',TextSelector(key='metaphone_repr')),
        ('soundex_ngrams',TextSelector(key='soundex_repr')),
        ('length_repr_ngrams',TextSelector(key='length_repr')),
        ('punctC_repr_ngrams',TextSelector(key='punctC_repr')),
        ('shape_repr_ngrams',TextSelector(key='shape_repr')),
        ('vowel_repr_ngrams',TextSelector(key='vowel_repr')),
        ('vowel_repr_word_ngrams',vowel_repr_word_ngrams),
        ('alliteration_repr',alliteration_repr),
        ('ner_repr',ner_repr),
        ('pos_repr_ngrams',TextSelector(key='pos_repr')),
        ('syllable_repr_ngrams',TextSelector(key='syllab_repr')),
    ]
    return feat_names

def write_grid_search_settings_to_pickle(settings_dict, path):
    path = path[:-10]
    with open(path+'_settings.p','wb') as outfile:
        pickle.dump(settings_dict,outfile)

feats = generate_feats_gender()
for i,f in enumerate(feats):
    print(i,f[0])
feat_names = return_feat_names()
for i,f in enumerate(feat_names):
    print(i,f[0])
```

```python
def generate_best_feats():
    with open(train_path[:-10]+'_settings.p','rb') as pickle_in:
        settings_dict = pickle.load(pickle_in)
    best_feats = []
    for key in settings_dict.keys():
        if settings_dict[key]['max']['settings']['Vectorizer'] == 'tfidf':
            vec = TfidfVectorizer(ngram_range=settings_dict[key]['max']['settings']['n-grams'],
                                             analyzer=settings_dict[key]['max']['settings']['Analyzer'],
                                             min_df=settings_dict[key]['max']['settings']['Min_df'],
                                             lowercase=False)
        best_feats.append((key,Pipeline([
            ('selector',settings_dict[key]['max']['settings']['Selector']),
            (settings_dict[key]['max']['settings']['Vectorizer'],vec)
        ])))
    print(best_feats[0][1][1].get_params())
    return best_feats
                

def optimize_best_feats(best_feats):
    changed = True
    for feat in best_feats:
        n_gram_range = feat[1][1].get_params()['ngram_range']
        start_i = n_gram_range[0]
        start_j = n_gram_range[1]
        analyz = feat[1][1].get_params()['analyzer']
        selector = feat[1][0]
        #print(selector)
        for name_vec in ['tfidf']:
            print(name_vec)
            for min_df in [5]:
                print(min_df,end=" ")
                if min_df == 10:
                    print()
                for i in range(start_i,start_i+2):
                    if i <= 0:
                        i = 1
                    for j in range(start_j,start_j+2):
                        if j < i:
                            j = i
                            if changed == True:
                                    with open(train_path[:-10]+'_settings.p','rb') as pickle_in:
                                        settings_dict = pickle.load(pickle_in)
                            changed = classify(feat,x_train,settings_dict,x_test,min_df,i,j,name_vec,selector,analyz)
    return settings_dict
                        
```

```python
def classify(feat,x_train_temp,settings_dict,x_test_temp,df,i,j,name_vec,selector,analyz):
    cls = LinearSVC()
    if name_vec == 'tfidf':
        vec = TfidfVectorizer(lowercase=False)
    else:
        vec = CountVectorizer(lowercase=False)
    pipe = Pipeline([
        ('selector',selector),
        ('vec',vec)
    ])
    pipe.set_params(vec__analyzer=analyz)
    pipe.set_params(vec__min_df=df)
    pipe.set_params(vec__ngram_range=(i,j))
    x_train_temp = ''
    x_test_temp = ''
    #x_test_temp = x_test.copy()
    x_train_temp = x_train.copy()
    #print(x_train.head()settings_dict[feat[0]+'_'+analyz[:4]]['tried'])
    settings = {'cls':'LinearSVC','Vectorizer':name_vec,'Analyzer':analyz,'Min_df':df,'n-grams':(i,j)}
    #settings_dict[feat[0]+'_'+analyz[:4]]['tried'].append(settings)
    #print(settings in settings_dict[feat[0]+'_'+analyz[:4]]['tried'])
    #print(settings_dict[feat[0]+'_'+analyz[:4]]['tried'])
    feat = [feat[0][:-5]]
    if settings not in settings_dict[feat[0]+'_'+analyz[:4]]['tried']:
        settings_dict[feat[0]+'_'+analyz[:4]]['tried'].append(settings)
        try:
            pipe.fit(x_train_temp)
            classifier = Pipeline([
                ('feats',pipe),
                ('cls',cls)
            ])
            #classifier.fit(x_train_temp, y_train)
            scores = cross_validate(classifier, x_train_temp, y_train, cv=9, scoring=("accuracy","f1_macro"))
            f1 = round(sum(scores['test_f1_macro']) / len(scores['test_f1_macro']),3)
            if f1 > settings_dict[feat[0]+'_'+analyz[:4]]['max']['f1']:
                print('NEW BEST:',f1)
                settings_dict[feat[0]+'_'+analyz[:4]]['max']['f1'] = f1
                settings_dict[feat[0]+'_'+analyz[:4]]['max']['settings'] = settings
        
        except:
            pass
        write_grid_search_settings_to_pickle(settings_dict,train_path)
        return True
    return False


def best_parameters_per_feature(feats,x_train,x_test):
    f1_word = 0
    changed = True
    for feat in feats:
        max_f1 = 0
        max_f1_char = 0
        selector = feat[1]
        print("\n\n"+str(selector))
        for name_vec in ['tfidf']:
            print(name_vec)
            for df in [5]:
                print("df:",df)
                for analyz in ['word','char_wb','char']:
                    print(analyz)
                    for i in range(1,10,1):
                        print(i, end =" ")
                        for j in range(i,i+6,1):
                            if j < 15:
                                #print(" [{}: ({},{})] ".format(analyz,i,j),end=" ")
                                if changed == True:
                                    with open(train_path[:-10]+'_settings.p','rb') as pickle_in:
                                        settings_dict = pickle.load(pickle_in)
                                changed = classify(feat,x_train,settings_dict,x_test,df,i,j,name_vec,selector,analyz)
                                
                                     
    return settings_dict
```

```python
def try_vis(conf_matrix, y_test, y_pred):
    """ Try to visualize confusion matrix """
    try:
        vis(conf_matrix, sorted(list(set(y_test))))
    except:
        try:
            vis(conf_matrix, sorted(list(set(y_pred))))
        except:
            pass

def vis(conf_mat, labels):
    '''Visualise the Confusion matrix'''
    df_cm = pd.DataFrame(conf_mat, index=[i for i in labels], columns=[i for i in labels])
    plt.figure(figsize=(10,7))
    plt.title('Confusion matrix')
    sn.heatmap(df_cm, annot=True, fmt='g')
    plt.show()




def most_informative_features(x_train, y_train, x_test, y_test, ngram_min, ngram_max, analyz):
    """ Gives a list of the coefficient per word, all of them, and gives  """


    vec = TfidfVectorizer(ngram_range=(ngram_min, ngram_max), analyzer=analyz)
    x_train = [row['lyrics'] for index,row in x_train.iterrows()] 
    x_train = vec.fit_transform(x_train)
    classifier = LinearSVC(C=200)
    classifier.fit(x_train, y_train)
    #classifier.fit(x_train, y_train)
    #classifier.score(x, y)

    feature_names = vec.get_feature_names()
    coefs_with_fns = sorted(zip(classifier.coef_[0], feature_names))

    df = pd.DataFrame(coefs_with_fns)
    df.columns = 'coefficient', 'word'
    df.sort_values(by='coefficient')
    print(df)

    plot_coefficients(classifier, feature_names)

def plot_coefficients(classifier, feature_names, top_features=20):
    """To plot the top X features"""
    coef = classifier[1].coef_.ravel()
    top_positive_coefficients = np.argsort(coef)[-top_features:]
    top_negative_coefficients = np.argsort(coef)[:top_features]
    top_coefficients = np.hstack([top_negative_coefficients, top_positive_coefficients])

    # create plot
    plt.figure(figsize=(17, 15))
    #plt.title('Top {0} features for {1}'.format(top_features))
    plt.xlabel('Feature')
    plt.ylabel('Coefficient')
    plt.ylim(-6, 6)
    colors = []
    for c in coef[top_coefficients]:
        if c < 0:
            colors.append('grey')
        else:
            colors.append('black')

    plt.bar(np.arange(2 * top_features), coef[top_coefficients], color=colors)
    feature_names = np.array(feature_names)

    fet_list = (feature_names[top_coefficients].tolist()) # for when you want translation
    fet_names = feature_names[top_coefficients]

    plt.xticks(np.arange(1, 1 + 2 * top_features), fet_names, rotation=60)
    plt.show()
    
def print_top10(feats,x_train,y_train,x_test,_test):
    """Prints features with the highest coefficient values, per class"""
    #x_train = [row['lyrics'] for index,row in x_train.iterrows()] 
    vec = TfidfVectorizer(ngram_range=(2, 6), analyzer='char_wb')
    #vec = TfidfVectorizer(ngram_range=(2, 4), analyzer='word')
    vec = TfidfVectorizer(ngram_range=(1, 3), analyzer='word',stop_words='english')
    #x_train = vec.fit_transform(x_train)
    cls = LinearSVC()
    #cls = MultinomialNB()
    
    
    #for feat in feats[5:10]:
    
    
    
   # max_f1 = 0
   # for i in range(1,5):
   #     for j in range(i,i+5):
   #         print(i,j)
    for i in [1]:
        print(i)
        cls = LinearSVC(C=200)
        feat = ('metaphone', Pipeline([
        ('selector',TextSelector(key='ner_repr')),
        ('tfidf',TfidfVectorizer(ngram_range=(1, 2), analyzer='word', min_df=5,lowercase=False))
        ]))


                #feats=[1,feats[2]]
        feats=[1,feat]
        print(feats[1][0])
        feats[1][1].fit_transform(x_train)
        classifier = Pipeline([
                ('feats', feats[1][1]),
                ('cls', cls)
            ])

        scores = cross_validate(classifier, x_train, y_train, cv=9, scoring=("accuracy","f1_macro"))
        print("\nAverage macro f1 score with K-Vold (k=10): {0}\n ".format\
              (round(sum(scores['test_f1_macro']) / len(scores['test_f1_macro']),3)))
        #classifier = MultinomialNB()
        #classifier.fit(x_train, y_train)
        
        #y_pred = classifier.predict(x_test)
        #f1 = f1_score(y_test,y_pred,average='macro')
        #print(f1_score(y_test,y_pred,average='macro'))
           
    print(classification_report(y_test,y_pred))
    f1 = f1_score(y_test,y_pred,average='macro')
    class_labels = classifier.classes_
    feature_names = feats[1][1][1].get_feature_names() #feats[1][1][1] is the tfidf vecorizer

    feature_names = feats[1][1][1].get_feature_names()
    coefs_with_fns = sorted(zip(classifier[1].coef_[0], feature_names))

    plot_coefficients(classifier,feature_names)
    
    for i, class_label in enumerate(class_labels):
        top10 = np.argsort(classifier[1].coef_[i])[-10:] # classifier[1] is the cls
        features = [feature_names[j] for j in top10]
        print("{0}\t{1}".format(class_label, "\t".join(features)))
```
